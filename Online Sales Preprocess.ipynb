```python
df['ShippingCost'] = df['ShippingCost'].fillna(0)
df['WarehouseLocation'] = df['WarehouseLocation'].fillna('No Warehouse')
df.drop(columns=['CustomerID'], inplace=True)
```

```python
df.info()
```

```python
df[df['Quantity'] < 0]
```

```python
df[df['UnitPrice'] < 0]
```

```python
df['UnitPrice'] = df['UnitPrice'].apply(lambda x: abs(x))
df['Quantity'] = df['Quantity'].apply(lambda x: abs(x))
```

```python
df['SalesAmount'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df['TotalDiscounts'] = df['SalesAmount'] * df['Discount']
df['TotalSales'] = df['SalesAmount'] - df['TotalDiscounts'] + df['ShippingCost']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year
df['Quarter'] = df['InvoiceDate'].dt.quarter
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
```
